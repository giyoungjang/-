- 적용할 함수를 리스트에 담아두고 각각의 문자열에 적용할 수 있다

```python
states = ['   Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda', 'south   carolina##', 'West virginia?']

def clean_strings(strings):
    result = []
    for value in strings:
        value = value.strip()
        value = re.sub('[!#?]', '', value)
        value = value.title()
        result.append(value)
    return result
```

```python
clean_strings(states)
```

```python
def remove_punctuation(value):
    return re.sub('[!#?]', '', value)

clean_ops = [str.strip, remove_punctuation, str.title]

def clean_strings(strings, ops):
    result = []
    for value in strings:
        for function in ops:
            value = function(value)
        result.append(value)
    return result
             
```

```python
clean_strings(states, clean_ops)
```

## numpy 배열 속성


- a.shape: 배열의 형태를 나타내는 튜플입니다. 예를 들어, (3, 4)는 3행 4열의 2차원 배열을 의미합니다.
- a.ndim: 배열의 차원(축의 수)을 나타내는 정수입니다. 예를 들어, 1차원 배열의 경우 1, 2차원 배열의 경우 2를 반환합니다.
- a.dtype: 배열의 데이터 타입을 나타냅니다. 예를 들어, int64, float32 등의 NumPy 데이터 타입 객체입니다.
- a.size: 배열의 전체 요소 수를 나타내는 정수입니다. 예를 들어, 3행 4열의 배열의 경우 12를 반환합니다.
- a.nbytes: 배열의 전체 메모리 크기를 바이트 단위로 나타내는 정수입니다. 일반적으로 a.size * a.itemsize와 동일합니다.
- a.itemsize: 배열의 각 요소의 메모리 크기를 바이트 단위로 나타내는 정수입니다. 예를 들어, int64의 경우 8, float32의 경우 4를 반환합니다.
- a.dtype: 배열 a의 데이터 타입을 나타냅니다. 예를 들어, int64, float32 등의 NumPy 데이터 타입 객체입니다.
-
- np.array(): 리스트나 튜플 등의 데이터를 입력으로 받아 NumPy 배열을 생성하는 함수입니다. 예를 들어, np.array([1, 2, 3])은 [1, 2, 3]이라는 1차원 NumPy 배열을 만듭니다.
- np.arange(): 일정한 간격의 값을 가진 NumPy 배열을 생성하는 함수입니다. np.arange(start, stop, step) 형태로 사용하며, start부터 stop까지 step 간격으로 값을 생성합니다. 예를 들어, np.arange(0, 10, 2)는 [0, 2, 4, 6, 8]이라는 1차원 NumPy 배열을 만듭니다.
- a.reshape(): 배열 a의 형태를 변경하는 함수입니다. 새로운 형태의 튜플을 입력으로 받아 배열의 형태를 변경합니다. 예를 들어, (3, 4) 형태의 2차원 배열 a를 a.reshape(4, 3)로 변경하면, (4, 3) 형태의 2차원 배열로 변경됩니다. 단, 변경하려는 형태의 요소 수와 원래 배열의 요소 수가 동일해야 합니다.

```python
position = 0
walk = [position]
steps = 1000
for i in range(steps):
    step = 1 if random.randint(0,1) else -1
    position += step
    walk.append(position)
walk[-10:]
plt.plot(walk[:1000])
```

## key 와 values 바꾸기  3가지 방법
 -   if type(values)==list:    
  -  if isinstance(values, list):   
   - if isinstance(j, typing.Hashalbe):


```python
di1 =  {"apple":["사과", '애플'], 
           "banana":["바나나", '빠나나'], 
           "cherry":"체리"}
di2 = {}
for key,values in di1.items():
    if type(values)==list:    #  if isinstance(values, list):   # if isinstance(j, typing.Hashalbe):
        for value in values:
            di2[value] = key
            
    else:
        di2[values] = key
di2
```
